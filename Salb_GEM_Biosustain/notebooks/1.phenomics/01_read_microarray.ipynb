```python
%%bash

CUR_DIR=`pwd`
DATA_DIR=$CUR_DIR"/../../data/raw/biolog/streptomyces_albus_J1074/run_01"
STRAIN="J1074"
MICROARRAY_DATA=$DATA_DIR"/samples_albus_J1074.csv"

# Initialize the project
dape init
# add my organism using the ID MyOrg
dape add $STRAIN

# add the phenomic experiment, Phenotype microarray data
dphenome add $MICROARRAY_DATA $STRAIN
# perform control subtraction
dphenome zero
```

```python
%%bash

# calculate the growth parameters and perform the clusterization, 
dphenome start -e -g
```

```python
%%bash

# based on elbow test, it shows that 10 clusters is the best
dphenome start -f -g -n 10
```

```python
%%bash

# plot the growth curves
dphenome plot
# plot the phenomic ring
dphenome rings
# restore the purged replicas
dphenome stats
dphenome export
```
